### matplotlib 基础

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
sinx = np.sin(x)
cosx = np.cos(x)
```

```python
plt.plot(x, sinx)
plt.plot(x, cosx)
plt.show()
```

```python
plt.plot(x, sinx, label="sin(x)")
plt.plot(x, cosx, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title("math func")
# 对中文支持不太好
plt.legend()
plt.show()
```

### Scatter Plot --- 散点图

```python
plt.scatter(x, sinx)
plt.show()
```

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.scatter(x, y, alpha=0.3)# alpha 不透明度
plt.show()
```
